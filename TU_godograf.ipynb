Годограф для функции x^2-3x+1

```python

import numpy as np
import matplotlib.pyplot as plt

# Диапазон значений w
w = np.linspace(-10, 10, 400)

# Вычисление действительной и мнимой частей функции
u = -w**2 + 1
v = -3*w

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(u, v, label=r'$f(i\omega) = -\omega^2 - 3i\omega$ +1')
plt.xlabel(r'Re{$f(i\omega)$} = $-\omega^2 + 1$')
plt.ylabel(r'Im{$f(i\omega)$} = $-3*\omega$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
# plt.show()
plt.savefig('Godograf2.eps')
```
